```python
import polars as pl
```

```python
dataframe = input("Enter the filename here: ")
```

```python

def coordinate_extractor(dataframe):
  '''Extracts the genomic co-ordinates'''
  #Load the file
  df = pl.read_csv(dataframe)

  #Extarct the coordinates
  df = df.with_columns(
       Coordinate = pl.col('Mutation').str.extract(r"^(\d+)", 1).cast(pl.Int64))
  return df
new_dataframe = coordinate_extractor(dataframe)
new_dataframe.write_csv('coordinates_table.csv')
```
